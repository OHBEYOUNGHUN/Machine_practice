```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests


def get_data(symbol):
    url = 'https://finance.naver.com/item/sise.nhn?code={}'.format(symbol)
    with urlopen(url) as doc:
        soup = BeautifulSoup(doc, "lxml", from_encoding="euc-kr")
        cur_price = soup.find('strong', id='_nowVal') 
        cur_rate = soup.find('strong', id='_rate')  # ②
        stock = soup.find('title')  # ③
        stock_name = stock.text.split(':')[0].strip()  # ④
        return cur_price.text, cur_rate.text.strip(), stock_name
print(get_data('005930'))
```

```python
#셀트리온 구식버전
#위에 데이터프레임 활용할것
from matplotlib import pyplot as plt
from matplotlib import dates as mdates
from mpl_finance import candlestick_ohlc
from datetime import datetime

#차트 출력으르 위해 위 데이터프레임 가공
import pandas as pd
from matplotlib import pyplot as plt
df = df.dropna()
df = df.iloc[0:30] #1년 데이터는 너무 많으므로 30개만 쓰자
df = df.sort_values(by='날짜') #오름차순으로 변경
for idx in range(0,len(df)):
    dt = datetime.strptime(df['날짜'].values[idx], '%Y.%m.%d').date()
    #날짜 칼럼의 %Y.%m.%d 형식을 datetime형식으로 변환한다 
    df['날짜'].values[idx] = mdates.date2num(dt)
    #위에 datetime형을  다시 float형으로 변환한다
ohlc = df[['날짜','시가','고가','저가','종가']]#ohlc라는 인덱스를가진 데이터프레임을 새로 만든다

#mpl_finance로 캔들 차트 만들자
plt.figure(figsize = (9,6))
ax = plt.subplot(1,1,1)
plt.title('Celtrion (mpl_finance candle stick)')
candlestick_ohlc(ax, ohlc.values, width = 0.8, colorup= 'red', colordown = 'blue')
#candlestick_ohlc 자체가 함수다
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
#x축 레이블이 숫자야,'%Y-%m-%d' 형식으로 문자열 변환해주자
plt.xticks(rotation = 45)
plt.grid(color = 'gray', linestyle = '--')
plt.show()
```

```python
#신형 그래프
#맨뒤 페이지 숫자 구하기 아까한거
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import mplfinance as mpf


url = 'https://finance.naver.com/item/sise_day.nhn?code=068270&page=1'
with urlopen(url) as doc:
    html = BeautifulSoup(requests.get(url, headers = {'User-agent': 'Mozilla/5.0'}).text,'lxml') 
    #뷰티플 수프로 url넘겨주고 웹페이지 파싱할 방식을 써준다
    #코드 책 참고하지 말고 혼자 찾자
    #requests 안써줘서 못받았었다, headers 기억하자
    pgrr = html.find('td', class_='pgRR') #class_로 쓰는 이유는 햇갈리지말라고
    s= str(pgrr.a['href']).split('=') #/item/sise_day.nhn?code=068270&page=384를 분리한것
    #분리한거에서 마지막페이지를 뽑아내자
    last_page = s[-1]
    
#전체페이지읽기 아까한거
df = pd.DataFrame() #일별 시세를 저 장 할 데이터 프레임
siese_url = 'https://finance.naver.com/item/sise_day.nhn?code=068270'

for page in range(1, int(last_page)+1): #1페이지부터 last_page까지 반복
    page_url = '{}&page = {}'.format(siese_url, page)
    df = df.append(pd.read_html(requests.get(page_url, headers = {'User-agent': 'Mozilla/5.0'}).text)[0])#읽은 한페이지 분량의 데이터 프레임을 df에 추가
#데이터 프레임 가공
df = df.dropna()
df = df.iloc[0:30] #1년 데이터는 너무 많으므로 30개만 쓰자
df = df.rename(columns = {'날짜':'Date','시가':'Open','고가':'High','저가':'Low','종가':'Close',
                         '거래량':'Volume'}) #한글컬럼명 영문으로 바꿈
df = df.sort_values(by = 'Date')
df.index = pd.to_datetime(df.Date) #Date칼럼을 DatetimeIndex형으로 변경한 후 인덱스로 설정
df = df[['Open','High','Low','Close','Volume']] #이것만 가져오자

#mplfinance
mpf.plot(df, title = 'Celltrion Candle chart', type = 'candle')
```

```python
#mplfinance
mpf.plot(df, title = 'Celltrion Candle chart', type = 'ohlc')
```

```python
kwargs = dict(title = 'Celltrion customized chart', type= 'candle', mav=(2,4,6),
             volume = True, ylabel='ohlc candles')
#kwargs 는 keyword argument의 약자이며, mpf.plot함수를 호출 할때 쓰이는 여러 인수를 담는 딕셔너리
mc = mpf.make_marketcolors(up='r', down='b', inherit=True)
#마켓색상은 스타일을 지정하는 필수객체 inherit = 관련색상도 비슷하게할까요? 이뜻
s = mpf.make_mpf_style(marketcolors = mc) #마켓색상을 인수로 넘겨줌
mpf.plot(df, **kwargs, style = s)
```
